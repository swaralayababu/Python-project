#4.Find out under which team and position, spending in terms of salary is high.

```python
df.sort_values(by='Salary', inplace=True)
df
```

```python
df2 = df[0:30]
df2
```

```python
Top_spending_team_position = df2.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print(f'Team and Position with High Salary Spending: {Top_spending_team_position}')
team_position_salary = df2.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Salary', hue='Position', data=team_position_salary,)
plt.suptitle("ABC COMPANY PVT LTD")
plt.title('Salary Spending by Team and Position')
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Total Salary Spending')
plt.show()
```

### Based on the analysis, It's evident that the Los Angeles Lakers, Cleveland Cavaliers, and New York Knicks employees in the position of SF have the highest salaries.

#5.Find if there is any correlation between age and salary , represent it visually.

```python
plt.figure(figsize=(10, 6))
plt.scatter(x='Age', y='Salary', data=df,color='r', edgecolor='black', alpha=0.8)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.suptitle("ABC COMPANY LTD")
plt.title('Violin Plot of Age vs Salary')
plt.show()
```

### This dataset included 57 ABC Company workers' personal data. The age of the employee and their pay were found to positively correlate with one another. Salary was plotted on the x-axis and age was placed on the y-axis in this scatter plot. It's evident that as people age, their salaries rise as well. However, after the age of 35, pay gradually declines.
